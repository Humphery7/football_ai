```python
data_f = data_final.sort_values('date',ascending=True).reset_index(drop=True)
data_f
```

```python
data_f.drop(columns=['xg','xga','venue'],inplace=True)
```

```python
train = data_f[data_f["date"] < '2022-01-01']

test = data_f[data_f["date"] > '2022-01-01']
```

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score

train_date = train['date']
test_date = test['date']

train_f = train.drop(columns='date')
test_f = test.drop(columns='date')
```

```python
train_f
```

```python
train_predictors = train_f.drop(columns='result')
train_target = train_f['result']

test_predictors = test_f.drop(columns='result')
test_target = test_f['result']
```

```python
grad_boost = GradientBoostingClassifier()

model = grad_boost.fit(train_predictors, train_target)

pred = model.predict(train_predictors)

acc_ = accuracy_score(train_target,pred)
f1_ = f1_score(train_target,pred,average='weighted')

print(f' accuracy_score: {acc_}')
print(f' f1_score: {f1_}')
```

```python
pred_test = model.predict(test_predictors)

acc_test = accuracy_score(test_target,pred_test)
f1_test = f1_score(test_target,pred_test,average='weighted')

print(f' accuracy_score: {acc_test}')
print(f' f1_score: {f1_test}')
```
